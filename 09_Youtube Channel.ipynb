# Youtube Channel

The YouTube channel dataset used in this analysis was obtained from Kaggle. It contains information about multiple YouTube channels, including channel names, video uploads, number of subscribers, and total video views. 

### Data Exploration
The analysis begins with Exploratory Data Analysis (EDA) to better understand the dataset. EDA tasks include:

- Data visualization: Creating visualizations like bar charts, line plots, and scatter plots to understand the distribution and relationships between variables.
- Statistical summaries: Calculating descriptive statistics, such as mean, median, and mode, to summarize numerical variables.
- Channel performance analysis: Analyzing the number of video uploads, subscribers, and video views to identify popular channels and trends.
- Audience engagement analysis: Examining the relationship between the number of subscribers, video views, and audience engagement metrics.
- Data cleansing: Handling missing values, outliers, and data inconsistencies to ensure data quality.






### Insights and Findings
The EDA of the YouTube channel dataset yielded the following key insights:

- Identification of popular channels based on the number of subscribers and video views.
- Understanding the relationship between channel performance metrics (video uploads, subscribers, video views) and audience engagement.
- Analysis of content trends and their impact on channel growth.<br>
The findings provide valuable insights for content creators, marketers, and YouTube channel owners to understand channel performance, audience engagement, and inform content strategies.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import  numpy as np
```

```python
df=pd.read_csv("Datasets/Youtube Channel/top-5000-youtube-channels.csv")
df.head()
```

```python
df.head(-5)
```

```python
df.tail(-5)
```

```python
print("Number of rows",df.shape[0])
print("Number of columns",df.shape[1])
```

```python
df.info()
```

```python
pd.options.display.float_format='{:.2f}'.format
```

```python
df.describe(include='all')
```

```python
df.head(20)
```

```python
df=df.replace('--',np.nan,regex=True)
```

```python
df.head(20)
```

```python
df.isnull().sum()
```

```python
per=df.isnull().sum()*100/len(df)
per
```

```python
sns.heatmap(df.isnull())
```

```python
df.dropna(axis=0,inplace=True)
```

```python
sns.heatmap(df.isnull())
```

```python
df["Rank"]=df["Rank"].str[0:-2]
```

```python
df.head()
```

```python
df["Rank"]=df["Rank"].str.replace(",","").astype('int')
```

```python
df.tail()
```

```python
df.dtypes
```

```python
df['Video Uploads']=df['Video Uploads'].astype("int")
df['Subscribers']=df['Subscribers'].astype("int")
```

```python
df.dtypes
```

```python
df['Grade'].unique()
```

```python
df['Grade']=df['Grade'].map({'A++ ':5, 'A+ ':4, 'A ':3, 'A- ':1, 'B+ ':0})
```

```python
df.head()
```

```python
df["Average_view"]=df['Video views']/df['Video Uploads']
```

```python
df.head()
```

```python
df.sort_values(by='Video Uploads',ascending=False).head()
```

```python
df.corr()
```

```python
sns.barplot(x='Grade',y='Video Uploads',data=df)
plt.show()
```

```python
sns.barplot(x='Grade',y='Average_view',data=df)
plt.show()
```

```python
sns.barplot(x='Grade',y='Subscribers',data=df)
plt.show()
```

```python
df.groupby('Grade').mean(numeric_only=True)
```

```python
sns.barplot(x='Grade',y='Video views',data=df)
plt.show()
```
