### Data Exploration
The analysis begins with Exploratory Data Analysis (EDA) to better understand the dataset. EDA tasks include:

- Data visualization: Creating meaningful plots and charts to visualize distributions, relationships, and patterns in the data.
- Statistical summaries: Calculating descriptive statistics, such as mean, median, and standard deviation, to summarize the dataset's numerical variables.
- Correlation analysis: Examining correlations between variables to identify potential relationships and dependencies.
- Data cleansing: Handling missing values, outliers, and data inconsistencies to ensure data quality.
